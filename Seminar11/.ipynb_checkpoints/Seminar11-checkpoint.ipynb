```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
def f(x):
    return -12 * x ** 4 * np.sin(np.cos(x)) - 18 * x ** 3 + 5 * x ** 2 + 10 * x - 30
```

```python
# 1. Определить корни
```

```python
roots = fsolve(f, [-2, -1, 0, 1, 2])
print(f"Корни функции: {roots}")
```

```python
# 2. Найти интервалы, на которых функция возрастает
```

```python
increasing_intervals = []
for i in range(len(roots) + 1):
    if i == 0:
        interval = [float('-inf'), roots[i]]
    elif i == len(roots):
        interval = [roots[i - 1], float('inf')]
    else:
        interval = [roots[i - 1], roots[i]]
    if f(interval[0]) < f(interval[1]):
        increasing_intervals.append(interval)
print(f"Интервалы, на которых функция возрастает: {increasing_intervals}")
```

```python
# 3. Найти интервалы, на которых функция убывает
```

```python
decreasing_intervals = []
for i in range(len(roots) + 1):
    if i == 0:
        interval = [float('-inf'), roots[i]]
    elif i == len(roots):
        interval = [roots[i - 1], float('inf')]
    else:
        interval = [roots[i - 1], roots[i]]
    if f(interval[0]) > f(interval[1]):
        decreasing_intervals.append(interval)
print(f"Интервалы, на которых функция убывает: {decreasing_intervals}")
```

```python
# 4. Построить график
```

```python
x = np.linspace(-5, 5, 1000)
y = f(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции f(x)')
plt.grid()
plt.show()
```

```python
# 5. Вычислить вершину
```

```python
from scipy.optimize import minimize_scalar
res = minimize_scalar(f)
print(f"Вершина функции: ({res.x}, {res.fun})")
```

```python
# 6. Определить промежутки, на котором f > 0
```

```python
positive_intervals = []
for i in range(len(roots) + 1):
    if i == 0:
        interval = [float('-inf'), roots[i]]
    elif i == len(roots):
        interval = [roots[i - 1], float('inf')]
    else:
        interval = [roots[i - 1], roots[i]]
    if f(interval[0]) > 0:
        positive_intervals.append(interval)
print(f"Промежутки, на которых f > 0: {positive_intervals}")
```

```python
# 7. Определить промежутки, на котором f < 0
```

```python
negative_intervals = []
for i in range(len(roots) + 1):
    if i == 0:
        interval = [float('-inf'), roots[i]]
    elif i == len(roots):
        interval = [roots[i - 1], float('inf')]
    else:
        interval = [roots[i - 1], roots[i]]
    if f(interval[0]) < 0:
        negative_intervals.append(interval)
print(f"Промежутки, на которых f < 0: {negative_intervals}")
```
